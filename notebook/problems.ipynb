## Exercise 1 

Two dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO

# import neccessary libraries
import random as random

# define our variables
number_of_simulations = 250000
sum_of_dice = 0
was_greater_than_7 = 0
was_even_number = 0

# loop through and add to our counters
for i in range(number_of_simulations):
    sum_of_dice = random.randint(1,6) + random.randint(1,6)
    if sum_of_dice > 7:
        was_greater_than_7 += 1
    if sum_of_dice % 2 == 0:
        was_even_number += 1
        
# divide our counters by our number of iterations (1000)
print(f"Based on {number_of_simulations} iterations, the estimated probablity of rolling a number greater than 7 is: {(was_greater_than_7 / number_of_simulations) * 100:.2f}%")
print(f"Based on {number_of_simulations} iterations, the estimated probablity of rolling an even is: {(was_even_number / number_of_simulations) * 100:.2f}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
number_of_simulations = 10000
ball_box = {}
five_ball_selection = []
three_white_two_red = 0
all_same_color = 0

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
            
# TODO
for i in range(number_of_simulations):
    for j in range(5):
        five_ball_selection.append(ball_box[random.randint(0,59)])
    # now we want to compare five_ball_selection to the two situations we're looking for
    # if our 5 random selections matches one of our situations, we want to add 1 to the respective counter
    if five_ball_selection.count('White') == 3 and five_ball_selection.count('Red') == 2:
        three_white_two_red += 1
    if len(set(five_ball_selection)) == 1:
        all_same_color += 1
    five_ball_selection.clear()    

print(f"Based on {number_of_simulations} iterations, the estimated probablity of getting three white and two red balls is: {(three_white_two_red / number_of_simulations) * 100:.2f}%")
print(f"Based on {number_of_simulations} iterations, the estimated probablity of all balls being the same color is: {(all_same_color / number_of_simulations) * 100:.2f}%")

```
